```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# 数据导入

```python
bank = pd.read_csv('./data/bank-additional-full.csv', sep=';')
bank.head()
```

```python
bank.columns
```

# 数据分析与处理

## 筛选部分列数据

```python
columns = ['age', 'job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome', 'pdays', 'duration', 'y']
data = bank[columns]
data.head()
```

## 处理缺失值

```python
all = data.shape[0]
print('count: {}'.format(all))
flag = True
for column in columns:
    if column != 'age':
        print(column + ":\n", data[column].unique())
        counts = data[column].value_counts()
        if column == 'contact':
            flag = False
        if flag:
            print('unknown: {}'.format(counts['unknown']))
        print("======================================")
```

## 营销结果映射

```python
data['y'] = data['y'].map({'yes': 1, 'no': 0})
data.head()
```

## 年龄处理

### age 分布分析

```python
plt.figure(figsize=(20,6))
sns.countplot(x='age', data=data)
```

* 从 age 图看出，银行客户的年龄分布数据并不均匀。下面将年龄数据离散化，分成多段

```python
print('Mean: ', round(bank['age'].mean(), 2))
print('STD: ', round(bank['age'].std(), 2))
```

```python
# 计算百分位数
print('Age 25%: ', bank['age'].quantile(q=0.25))
print('Age 50%: ', bank['age'].quantile(q=0.5))
print('Age 75%: ', bank['age'].quantile(q=0.75))
```

### 年龄分段并转为哑变量

```python
age_cut = pd.qcut(data['age'], 6)
```

```python
age_dummies = pd.get_dummies(age_cut, prefix='age')
age_dummies
```

```python
age_dummies.columns = ['age_'+str(x) for x in range(6)]
age_dummies
```

### 相关性分析

```python
plt.figure(figsize=(10,8))
sns.heatmap(age_dummies.join(data['y']).corr(), annot=True)
```

* 相关性不高，考虑舍弃

```python
final_data = data.copy()
final_data.drop(labels=['age'], axis='columns', inplace=True)
final_data = final_data.join(age_dummies)
final_data
```

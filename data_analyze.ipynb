1.7. Bank Client Categorical Treatment¶

* Jobs, Marital, Education, Default, Housing, Loan. Converting to continuous due the feature scaling will be apllyed later

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
bank_clients['job'] = labelencoder_X.fit_transform(bank_clients['job'])
bank_clients['marital']  = labelencoder_X.fit_transform(bank_clients['marital']) 
bank_clients['education'] = labelencoder_X.fit_transform(bank_clients['education']) 
bank_clients['default'] = labelencoder_X.fit_transform(bank_clients['default']) 
bank_clients['housing'] = labelencoder_X.fit_transform(bank_clients['housing']) 
bank_clients['loan'] = labelencoder_X.fit_transform(bank_clients['loan']) 
```

```python
bank_clients.head()
```

```python
# function to creat group of ages, this helps because we have 78 different values here
print('Age 25%: ', bank_clients['age'].quantile(q=0.25))
print('Age 50%: ', bank_clients['age'].quantile(q=0.5))
print('Age 75%: ', bank_clients['age'].quantile(q=0.75))

def age(data):
    data.loc[data['age'] <= 32, 'age'] = 1
    data.loc[(data['age'] > 32) & (data['age'] <= 47), 'age'] = 2
    data.loc[(data['age'] > 47) & (data['age'] <= 70), 'age'] = 3
    data.loc[(data['age'] > 70) & (data['age'] <= 98), 'age'] = 4
    return data
    

age(bank_clients)
```
